### WEB API

Há WEB API para tudo hoje em dia! Vamos explorar uma API de noticias, eles concetram dados de mais de 30000 sites de notícias e jornais do mundo todo.

- Crie sua própria chave de acesso: https://newsapi.org/
- E siga-os no Twitter: https://twitter.com/NewsAPIorg

```python
import requests
import pandas as pd
```

```python
#Obtenha sua própria chave no site da API
KEY="a471bfae6bea4870828911cc205e47e9"
q="Brazil"

URL = "http://newsapi.org/v2/everything?q=" + q + "&from=2020-03-01&sortBy=publishedAt&apiKey=" + KEY

response = requests.get(URL)
```

Para saber mais sobre a sintaxe de um URL ou URI:

- https://pt.wikipedia.org/wiki/URL
- https://www.w3schools.com/html/html_urlencode.asp

```python
pd.DataFrame([ [a['title'],a['description']] for a in response.json()['articles'] ], 
             columns=['title','description'])
```
