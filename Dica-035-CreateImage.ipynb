Criar e salvar imagem com lib Pillow (PIL):

````python
from PIL import Image, ImageDraw

# Create a new image with a white background
width, height = 600, 600
img = Image.new('RGB', (width, height), 'white')
draw = ImageDraw.Draw(img)

# Draw a blue square
square_size = 200
square_x = (width - square_size) // 2 - 150 # shifted left
square_y = (height - square_size) // 2
draw.rectangle([square_x, square_y, square_x + square_size, square_y + square_size], fill='blue')

# Draw a red circle
circle_radius = 100
circle_x = width // 2  # Center X
circle_y = height // 2  # Center Y
# shifted right
circle_center_x = circle_x + 150
draw.ellipse([circle_center_x - circle_radius, circle_y - circle_radius, circle_center_x + circle_radius, circle_y + circle_radius], fill='red')

# Draw a green triangle
triangle_points = [(300, 100), (200, 300), (400, 300)] # above the square and circle
draw.polygon(triangle_points, fill='green')

# Display the image
display(img)

# Save the image
img.save('i1.png')
````
