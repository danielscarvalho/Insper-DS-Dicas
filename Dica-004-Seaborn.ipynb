Importar as libs

```python
import seaborn as sns
import matplotlib.pyplot as plt

#exibe a saida do gráfico no próprio jupyter notebook 
%matplotlib inline   
sns.set(color_codes=True)
```

Exemplo 1: Tips

Carregar o dataset

```python
tips = sns.load_dataset("tips")
tips.sample(10)
```

Visualizar...

```python
sns.violinplot(x = "total_bill", data=tips)
plt.show()
```

```python
sns.violinplot(x = "tip", data=tips)
plt.show()
```

```python
sns.violinplot(x = "tip", y="total_bill", data=tips)
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker",
            kind="violin", split=True, data=tips);
```

```python
sns.set() #reset
sns.relplot(x="total_bill", 
            y="tip", 
            col="time",
            hue="smoker", 
            style="smoker", 
            size="size",
            data=tips);
```

Scatter plot grid

```python
view = sns.FacetGrid(tips, col="sex", 
                           row="smoker", 
                 margin_titles=True)

view.map(plt.scatter, "total_bill", "tip")

view.fig.suptitle("Tips Analisys")

plt.show()
```

```python
sns.catplot(x="day", y="total_bill", data=tips);
```

```python
sns.catplot(x="day", y="total_bill", 
            kind="box", 
            data=tips,
            hue="smoker");
```

Exemplo 2: Iris Species

Carregar dados

```python
iris = sns.load_dataset("iris")
iris.sample(10)
```

Visualizar...

```python
sns.swarmplot(x="species", y="petal_length", data=iris)
plt.show()
```

```python
sns.swarmplot(x="species", y="sepal_length", data=iris)
plt.show()
```

Example 3: Titanic

Carregar dados...

```python
titanic = sns.load_dataset("titanic")
titanic.sample(10)
```

Visualizar...

```python
sns.catplot("class", "survived", "sex", 
            data=titanic, 
            kind="bar", 
            palette="dark", 
            legend=False)

plt.show()
```

```python
sns.regplot("age", "fare", 
            data=titanic, 
            robust=True, 
            ci=None, 
            color="blue")

sns.despine();
```

```python
sns.catplot(x="pclass",
            y="survived",
            kind="point", 
            data=titanic, 
            palette="dark");
```

```python
g = sns.catplot(x="who", y="survived", col="class",
                data=titanic, saturation=.5,
                kind="bar", ci=None, aspect=.6)

(g.set_axis_labels("Category", "Survival Rate")
  .set_xticklabels(["Men", "Women", "Children"])
  .set_titles("{col_name} {col_var}")
  .set(ylim=(0, 1))
  .despine(left=True) )
```

```python
fg = sns.FacetGrid(titanic, hue="sex", aspect=3, palette="dark")
fg.map(sns.kdeplot, "age", shade=True)
fg.set(xlim=(0, 80));
```

```python
fg = sns.FacetGrid(titanic, hue="who", aspect=3, palette="dark")
fg.map(sns.kdeplot, "age", shade=True)
fg.set(xlim=(0, 80));
```

```python
sns.lmplot("age", "survived", titanic, 
           logistic=True, y_jitter=.05);
```
