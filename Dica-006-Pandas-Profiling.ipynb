### Pandas-Profiling

O Pandas Profiling é uma ferramenta sofisticada para identificar as características dos dados em um dataframe Pandas (Nutela). É interessante para que o cientista de dados tenha uma boa perspectiva da qualidade dos dados e quais os tipos de análise que podem ser feitas, quais amostras são interessantes, quais podem ser descartados e assim por diante...

Importação das bibliotecas

```python
import numpy as np
import pandas as pd
from pandas_profiling import ProfileReport
```

Carga de dados no dataframe Pandas, esta é a mesma "tabela" que utilizamos em uma dica anterior

```python
df_coronav = pd.read_csv('data/corona/COVID19_open_line_list.csv')
```

```python
df_coronav.sample(10)
```

```python
df_coronav.describe()
```

A ferramenta vai ler os dados do dataframe e gerar um relatório detalhado...

Isso demora, depende do volume de dados do seu dataframe...

```python
profile = ProfileReport(df_coronav, title='Pandas Profiling Report')
```

Salvando o relatório em HTML:

```python
profile.to_file(output_file="coronav_profile.html")
```
