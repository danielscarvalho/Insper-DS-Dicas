### Recurrence Plot

Equação a ser implementada em Python:

$R_{i,j}=\Theta(\varepsilon_i - \|\vec{x}_i - \vec{x}_j\|), \qquad 
\vec{x}_i \in \mathbb{R}^m, \quad i,j = 1,\ldots, N,$

Definição:

Traduzido do inglês > Na estatística descritiva e na teoria do caos, um gráfico de recorrência é um gráfico que mostra, para cada momento i no tempo, os momentos em que uma trajetória do espaço de fase visita aproximadamente a mesma área no espaço de fase que no tempo j. Em outras palavras... Wikipedia (inglês)

```python
#Fonte: https://laszukdawid.com/tag/recurrence-plot/

import pylab as plt
import numpy as np
 
def rec_plot(s, eps=None, steps=None):
    if eps==None: eps=0.01
    if steps==None: steps=10
    N = s.size
    S = np.repeat(s[None,:], N, axis=0)
    Z = np.floor(np.abs(S-S.T)/eps)
    Z[Z>steps] = steps
 
    return Z
```

Usando o Recurrence Plot para visualizar uma série temporal, Bitcon trades

Consumindo dados via WEB API (WEB Service)

```python
import requests

bitcoin_trades = requests.get("https://www.mercadobitcoin.net/api/BTC/trades/").json()
```

```python
bitcoin_trades
```

Preços das transações...

```python
[b["price"] for b in bitcoin_trades]
```

Visualização da série temporal

```python
plt.plot([b["price"] for b in bitcoin_trades])
plt.show()
```

Recurrence Plot

```python
s = np.array([b["price"] for b in bitcoin_trades])
plt.imshow(rec_plot(s))
plt.show()
```
