Podemos publicar nossos modelos na Internet como WEB Service, de forma que qualquer developer possa utiliza-los, independente da linguagem de programação que utilizam.

Para isso vamos publicar um projeto em nuvem, em uma máquina virtual (VM) Linux no Microsoft Azure. Mas o mesmo pode ser feito na AWS, GCP, Oracle Cloud, etc...

1. Criar uma conta de estudante no Azure, para uso de infra gratuito (use o e-mail de aluno do Insper):

https://azure.microsoft.com/pt-br/free/students
    

2. Criar uma máquina virtual com Linux Ubuntu, utilizar a configuração de máquina mais simples:

3. Liberar a porta do firewall 3000 padrão do Flask

Na nuvem da Microsoft Azure, no menu Networking (em Settings ao lado esquerdo), adicione a porta 5000 para acesso externo (inbound):

> Temos um passo-a-passo no material de aula...

4. Conectar no servidor virtual Linux (VM). O arquivo .pem é a chave fornecida ao criar a VM Linux (você deve ter feito download deste arquivo), buzz é o nome do usuário definido na criação da VM Linux, e o IP válido na Internet fornecido pelo Azure ao iniciar a máquina virtual.

```bash
ssh -i chave.pem buzz@20.122.135.238
```

5. Atualizar o sistema operacional (Linux) da VM em nuvem, pode demorar...

```bash
sudo apt update
```

6. Verificar a versão do Python

```bash
python3 --version
```

7. Verificar a versão do Git

```bash
git --version
```

8. Verificar a versão do Linux

```bash
uname -a
```

9. Instalar o Flask (biblioteca do Python)

```bash
sudo apt install python3-flask
```

10. Clonar o código do projeto que está no GitHub

```bash
git clone https://github.com/danielscarvalho/flask-local
```

11. Entrar na pasta do projeto

```bash
cd flask-local
```

12. Inicializar o Flask

```bash
sh ./run.sh
```
