### Imagens de 2 (x, y) dimensões em tons de cinza

Podemos facilmente visualizar uma matrix, no caso a seguir 3x4 como uma imagem...

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
a = np.array([[1, 2, 3, 4], 
              [5, 9, 9, 8], 
              [1, 1, 2, 2]])
````

Note que os valores mais altos são as cores mais claras na escala

````python
fig, ax = plt.subplots()

ax.imshow(a)

plt.show()
````

Agora vamos visualizar uma função 2D carteseana de seno e cosseno, que lembra uma caixa de ovos...

````python
# make data
X, Y = np.meshgrid(np.linspace(-3, 3, 16), np.linspace(-3, 3, 16))
Z = np.sin(X) * np.cos(Y)

# plota
fig, ax = plt.subplots()

ax.imshow(Z)

plt.show()
````

Novamente, mas agora com mais pixels... e com a palheta de cores 'hot'

````python
X, Y = np.meshgrid(np.linspace(-100, 100, 20), np.linspace(-100, 100, 20))
Z = np.sin(X) * np.cos(Y)

fig, ax = plt.subplots()

plt.set_cmap(plt.get_cmap('hot'))

ax.imshow(Z)

plt.show()
````

A mesma coisa mais agora visualizando a função trigonomética Sinc tanto em X quando tem Y.
Perceba que as cores (tons) são o valor do eixo Z:

````python
# make data
X, Y = np.meshgrid(np.linspace(-3, 3, 30), np.linspace(-3, 3, 30))
Z = np.sinc(X) * np.sinc(Y)

# plot
fig, ax = plt.subplots()

ax.imshow(Z)

plt.show()
````

Agora novamente visualizando valores aleatórios em uma mariz 50x50 e exibindo a escala de cores (color bar)

Agora vamos importar uma imagem da Internet (URL) e visualiza-la

````python
# importing modules 
import urllib.request 
from PIL import Image, ImageFilter


image_url="https://hips.hearstapps.com/hmg-prod/images/cute-photos-of-cats-looking-at-camera-1593184780.jpg?crop=0.6672958942897593xw:1xh;center,top&resize=980:*"

urllib.request.urlretrieve(image_url,"cat.jpeg") 
  
img = Image.open("cat.jpeg") 
img.show()
````

Podemos aplicar filtros (convolução) nas imagens também

````python
kernel = [9, 1, 9,
          1, 1, 1,
          9, 1, 9]

kernel_filter = ImageFilter.Kernel((3, 3), kernel, scale=11, offset=0)

# Apply filter on the image
cat_img_filter = cat_img.filter(kernel_filter)

# Save the resulting image
cat_img_filter.save("cat_filter.jpg")

fig, ax = plt.subplots()
ax.imshow(cat_img_filter)
plt.show()
````

Experimente outros filtros como:

````python
kernel = [0, 20, 0,
          20, 0, 20,
          0, 20, 0]

kernel_filter = ImageFilter.Kernel((3, 3), kernel, scale=11, offset=0)

# Apply filter on the image
cat_img_filter = cat_img.filter(kernel_filter)

# Save the resulting image
cat_img_filter.save("cat_filter2.jpg")

fig, ax = plt.subplots()
ax.imshow(cat_img_filter)
plt.show()
````

Entre diversas transformações de imagens, podemos converter a imagem para tons de cinza, assim passamos a ter uma matrix 2D

````python
cat_gray = cat_img.convert("L")

fig, ax = plt.subplots()
ax.imshow(cat_gray)
plt.show()
````
