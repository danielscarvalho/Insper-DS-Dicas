# Computando com Python e R no mesmo Jupyter Notebook

Em um ambiente Ubuntu Linux, no PC local... com o Anaconda instalado:

````python
!conda --version
````

Instalar a LIB "irkernel" para o Jupyter na linha de comando (terminal):

```bash
conda install -c r r-irkernel
```

Instarar o R no Linux pela linha de comando Bash (terminal):

```bash
sudo apt install R-base
```

Verificando a versão do R instalada no ambiente (SO):

````python
!R --version
````

No R instalar a LIB "IRkernel" também, no terminal:

1. Abrir o R
2. Executar o instalador da LIB

```bash
R
install.packages('IRkernel')

```

3. Sair do R com ```q()``` e pela linha de comando:

```bash
sudo R --slave -e 'quit("no", IRkernel::installspec(FALSE))'
```

Iniciar o Jupyter Notebook, e agora temos R como uma opção nos menus:






Com o ambiente configurado, agora é possível criar um Jupyter Notebook para programar em R:



Computando com o Python como de costume:

````python
import math

math.radians(33/33)
````
